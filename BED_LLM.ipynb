Статья исследует вопрос: можно ли научиться правильно выбирать запросы так, чтобы максимизировать эффективность LLM?
LLM плохи в поиске информации в проактивной манере. И неплохо было бы научить их отбирать вопросы на основе уже заданных.

Авторы предлагают использовать для ответа на этот вопрос Expected Information Gain (EIG) Estimator. А именно, не влезая во внутреннюю структуру LLM, сделать внешнюю систему, подбирающую вопросы таким образом, чтобы максимизировать получаемую информацию (и соответствующую функцию $\theta$).

их алгоритм выглядит следующим образом:

1. Generate Candidate Questions: Propose a candidate set of M diverse, multiple-choice
questions, X, by appropriate sampling of the LLM based on the current conversational
context (Section 3.1).
2. Compute EIG Estimator: For each candidate question x ∈ X cand, estimate EIG θ(x) as
described in in Section 3.2.
3. Select and Ask Optimal Question: Choose the question $x_t ∈ X$ that yields the
highest estimated EIG. Pose $x_t$ to the user, observe response $y_t$, and update the history,
$h_t = (h_{t−1}, (x_t, y_t))$.
4. Construct updated joint, $p(θ, y_{t+1}; h_t, x_{t+1}) = p_f (θ; h_t)p_{LLM} (y_{t+1}; [θ, x_{t+1}]),$ using
the new history (Section 3.3) and return to Step 1 (unless a termination criterion like
maximum number of questions has been achieved).

(здесь p - "распределения")

Авторы говорят, что получилось очень круто обучить модель для игры в "20 вопросов".

К сожалению, кода на гитхабе нет, реализации EIG я не нашел, а вчитываться и восстанавливать доказательства из области тервера сейчас не очень хочется, так что пусть остается на уровне идеи.
