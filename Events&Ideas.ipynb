Секция 3. ML in Advertising

Олеся Булгакова, X5

    - Как превратить рекламу в рекомендацию. Если рекламное место занято неинтересным пользователю товаром - бизнес и рекламодатель теряют деньги, а клиент - лояльность. Был разработан реранкер, подбирающий рекламу персонально под пользователя, как гибкий инструмент поиска компромисса между этими стремя сторонами. База: каждый третий товвар - рекламный. 1 шаг: Строим рекомендации с помощью bert+GBoosting. Если есть история - сравниваем эмбеддинги рекомендаций и рекламы, для холодного старта (новые товары) - по описаниям. Выручка выросла на 100%, однако показы рекламы упали на 90%. 2 шаг: персонализация, добавление признаков релевантности и прибыльности, исходя из бизнес-задачи. Падение показа рекламы уменьшилось. Шаг 3: управление рекомендацией в зависимости от типа рекламы. Была добавлена вероятность покупки и новая метрика для рекламодателя - рекламный доход за период на 1000 показов и производные от этой метрики. Замена бизнес-модели взаимодействия с рекламодателями привела к тому, что все три стороны оказались в плюсе. Шаг 4: переход к онлайн-пересчету рекомендаций.

Алексей Козлов, Авито

    - Полный обзор пайплайна ранжирования рекламных товаров Авито: бизнес метрики, модель аукциона, эмбеддинги, инференс...

Эдуард Колдунов, Digital Budget

    - Мультимодальная обработка банеров для их кластеризации и решения бизнес-задач. Решается задача дедупликации банеров хешированием и дообучением эмбеддеров.

Анастасия Никулина, Wildberries

    - Мультимодальная обработка банеров для их кластеризации и решения бизнес-задач. Решается задача оптимизации метрик и количества моделей. Кратко пересказана работа https://arxiv.org/pdf/1903.01057


Секция 4. Mathematics & ML/ML in DBMS

Никита Гущин, AIRI

    - Обратная дистилляция для ускорения диффузионных моделей. Разрабатывается техника, когда дистиллированная модель получает то же самое изображение, но за меньшее число шагов. Упомянуты эксперименты с текстовыми диффузионными моделями и построение диффузионных моделей на принципах электростатики (а не термодинамики, как сейчас). 

Алексей Шестов, Sber AI Lab https://t.me/botka_chronics

    - Выше много говорилось про эмбеддинги, здесь они анализируются с помощью топологии. Цель: оценка эмбеддингов для unsupervised реальных сценариев, причем заранее неизвестно, для каких. При фиксированном пороге соединением точек, расстояние между которыми не превосходит этого порога, формируется многообразие. Таким образом, при изменении порога возникают и исчезают разные компоненты связности и дырки в этих компонентах. Метрика основана на времени существования тех или иных топологических инвариантов. Метрика используется для оценки априорного качества эмбеддингов.
